```python
import matplotlib.pyplot as plt
import json

with open('E:\C\ICT\Graduation_Research_1\Code\MyAI_Recruitment\Data\DB\skill_db_taxonomy.json', 'r+') as f:
    data = json.load(f)
    
it_data = data.get("Information Technology", {})
subcategories = []
counts = []

for subcategory, subcategory_values in it_data.items():
    subcategories.append(subcategory)
    counts.append(len(subcategory_values))

# Vẽ biểu đồ cột với màu cam
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.4

bars = ax.bar(subcategories, counts, color='blue')  # Đặt màu cam cho tất cả các cột

# Thêm thông tin trục và tiêu đề
ax.set_xlabel('Subcategories')
ax.set_ylabel('Number of Items')
ax.set_title('Distribution of Subcategories in "Information Technology"')

# Xác định vị trí của ticks trên trục x
ax.set_xticks([i + bar_width/2 for i in range(len(subcategories))])
ax.set_xticklabels(subcategories, rotation=90, ha='right')

plt.tight_layout()
plt.show()
```
